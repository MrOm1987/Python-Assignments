```python
# Question: Find the union of two sets
set1 = {1, 2, 3, 4}
set2 = {4, 5, 6, 7, 8}
union_set = set1.union(set2)
print(union_set)
```

```python
# Question 1: Create a set of first 10 natural numbers

numbers = set(range(1, 11))
print(numbers)
```

```python
numbers = set()
for i in range(1, 11):
    numbers.add(i)
print(numbers)
```

```python
# Question 4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}
setA = {1, 2, 3, 4, 5} 
setB = {4, 5, 6, 7, 8}
intersec_set = setA.intersection(setB) #or set_diff = setA & setB
print(intersec_set)
```

```python
# Question 5: Find the difference between {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}
setA = {1, 2, 3, 4, 5} 
setB = {4, 5, 6, 7, 8}

set_diff = setA - setB #or set_diff = setA.difference(setB) 
print(set_diff)
```

```python
# Question 8: Create a set of vowels from string "hello world"
text = "hello world"
vowels_set = set()
for ch in text:
    if ch in 'aeiou':
        vowels_set.add(ch)
print(vowels_set)
```

```python
# Question 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set

lst = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
unique = list(set(lst))
print(unique)
```

```python
lst = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
unique = []
seen = set()
for num in lst:
    if num not in seen :
        seen.add(num)
        unique.append(num)
print(unique)
```

```python
# Question 10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}

setA = {1, 2, 3}
setB = {1, 2, 3, 4, 5, 6}

subset_s = setA.issubset(setB)
print(subset_s)
```
