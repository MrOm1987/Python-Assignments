```python
# Question: Find the key with maximum value in a dictionary
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'Diana': 95, 'Eve': 88}
max_key = max(scores, key= scores.get)
max_value = scores[max_key]
print(max_key)
print(max_value)
```

```python
# Question 9: Count frequency of each character in string "hello"
my_str = "hello"
frquncy_of_each_char = {}
for ch in my_str:
    if ch in frquncy_of_each_char:
        frquncy_of_each_char[ch] += 1
    else:
        frquncy_of_each_char[ch] = 1
print(frquncy_of_each_char)
```

```python
# Question 10: Merge two dictionaries {'a': 1, 'b': 2} and {'c': 3, 'd': 4}
my_dic1 = {'a': 1, 'b': 2}
my_dic2 = {'c': 3, 'd': 4}
my_dic1.update(my_dic2)
print(my_dic1)
```

```python
my_dic1 = {'a': 1, 'b': 2}
my_dic2 = {'c': 3, 'd': 4}
merged = {}

for key in my_dic1:
    merged[key] = my_dic1[key]
for key in my_dic2:
    merged[key] = my_dic2[key]
print(merged)
```
