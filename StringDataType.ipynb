```python
# Question 41: Check if string contains balanced parentheses: "((()))"
s = "((()))"
balance = 0
is_balanced = True
for char in s:
    if char == '(':
        balance += 1
    elif char == ')':
        balance -= 1
    if balance < 0:
        is_balanced = False
if balance != 0:
    is_balanced = False
print(is_balanced)
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
s1 = "programming"
s2 = "grammar"
longest = ""
for i in range(len(s1)):
    for j in range(i + 1, len(s1) + 1):
        substring = s1[i:j]
        if substring in s2 and len(substring) > len(longest):
            longest = substring
print(longest)
```

```python
# Question 46: Convert "hello world" to Pig Latin
sentence = "hello world"
vowel = "aeiou"
result = []
for word in sentence.lower().split():
    if word[0] in vowel :
        result.append(word + "way")
    else:
        result.append(word[1:]+word[0] + "ay")
pig_latin = ' '.join(result)  

print(pig_latin)
```

```python
# Question 49: Convert "hello world" to ROT13 encoding
str = "hello world"
result = []
for c in str :
    if 'a' <= c <= 'z':
        result.append(chr((ord(c) - ord('a') + 13) % 26 + ord('a') ))
    elif 'A' <= c <= 'Z' :
        result.append(chr((ord(c) - ord('A') +13) % 26 + ord('A')) )
    else :
        result.append(c)
roi13_code = ''.join(result)
print(roi13_code)
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"
number = "4532015112830366"
check_sum = 0
digits = [int(d)for d in number ]
reverse = digits[:: -1]
for i, d in enumerate(reverse):
    if i % 2 == 1 :
        d *= 2
        if d > 9:
            d -= 9
    check_sum += d
is_valid_cardnumber = check_sum % 10 == 0
print(is_valid_cardnumber)
        
```
