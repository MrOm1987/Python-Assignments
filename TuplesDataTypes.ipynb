```python
# Question: Find the sum and product of all elements in a tuple
numbers = (2, 4, 6, 8, 10)
sum_numbers = sum(numbers)
product = 1
for n in numbers:
    product *= n

print(sum_numbers)
print(product)
```

```python
# Question 4: Find the maximum value in tuple (23, 45, 12, 67, 34, 89, 56)
numbers_tuple = (23, 45, 12, 67, 34, 89, 56)
max_val_tuple = max(numbers_tuple)
print(max_val_tuple)
```

```python
numbers_tuple = (23, 45, 12, 67, 34, 89, 56)
max_val = numbers_tuple[0]
for num in numbers_tuple:
    if num > max_val:
        max_val = num
print(max_val)
```

```python
# Question 5: Count how many times 5 appears in (1, 5, 2, 5, 3, 5, 4, 5, 6)
t = (1, 5, 2, 5, 3, 5, 4, 5, 6)
count = 0
for num in t:
    if num == 5:
        count += 1
print(f"5 appears {count} times")
```

```python
# Question 10: Find the average of numbers in tuple (15, 23, 31, 42, 56, 78)
t = (15, 23, 31, 42, 56, 78)
avg = sum(t)/len(t)
print(avg)
```

```python
t = (15, 23, 31, 42, 56, 78)

total = 0
count = 0

for num in t :
    total += num
    count += 1
average = total / count
print(average)
```
