```python
# Question: Calculate the area of a circle with radius 5.5
import math
radius = 5.5
area = math.pi * radius ** 2
print(f"radius: {radius}")
print(f"ares of circle: {area:.2f}")
```

```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577
number = [3.14, 2.718, 1.618, 0.577]
avarage = sum(number) / len(number)
print(avarage)
```

```python
# Question 2: Convert 98.6 Fahrenheit to Celsius (F = C * 9/5 + 32)
fahrenheit = 98.6
celsius = (fahrenheit - 32) * 5/9
print(celsius)
```

```python
# Question 3: Calculate the compound interest on $1000 at 5.5% for 3 years
principal = 1000
intrest = 5.5/100  #convert intrest to decimal
time = 3
amount = principal * (1 + intrest) ** time
comp_intrest = amount - principal
print(f"{comp_intrest:.2f}")
```

```python
# Question 4: Find the hypotenuse of a right triangle with sides 3.5 and 4.2
a = 3.5
b = 4.2
hypotenuse = (a**2+b**2)**0.5 #hypotenus=square root of a + square root of b
print(hypotenuse) 
```

```python
# Question 5: Calculate the volume of a sphere with radius 7.8
# V=4/3*pi*r cube
import math
radius = 7.8
vol = 4/3*math.pi*radius ** 3
print(vol)
```

```python
# Question 6: Round 3.14159 to 3 decimal places
num=3.14159
print(f"{num:.3f}")
```

```python
# Question 8: Find the square root of 23.456
number = 23.456
square_root = number ** 0.5
print(square_root)
```

```python
# Question 9: Calculate the simple interest: Principal=2500, Rate=6.5%, Time=2.5 years
# simple interest = PTR/100
principal = 2500
rate = 6.5
time = 2.5
simpl_interest = (principal*time*rate)/100
print(simpl_interest)
```
