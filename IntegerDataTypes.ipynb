```python
# Question 6: Check if 97 is a prime number
number=97
is_prime = True
if number<=1:
    is_prime=False
else:
    for i in range(2,number):
        if number % i==0:
         is_prime=False
        break
    if is_prime:
        print(f"97 is prime")
    else:
        print(f"97 is not prime")
```

```python
# Question 9: Find the greatest common divisor (GCD) of 48 and 36
a=36
b=48
gcd=1
for i in range(1,min(a,b)+1):
    if a % i ==0 and b % i ==0:
        gcd = i
print(gcd)
```

```python
# Question 10: Calculate the sum of first 20 odd numbers
sum_odd = 0
num = 1
count = 0
while count < 20:
    sum_odd += num
    num += 2
    count += 1
print(sum_odd)
```
