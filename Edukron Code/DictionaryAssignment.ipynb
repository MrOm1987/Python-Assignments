```python
# Question 9: Count frequency of each character in string "hello"
my_str = "hello"
frquncy_of_each_char = {}
for ch in my_str:
    if ch in frquncy_of_each_char:
        frquncy_of_each_char[ch] += 1
    else:
        frquncy_of_each_char[ch] = 1
print(frquncy_of_each_char)
```

```python
# Question 35: Create a 4-level nested dictionary
nested_dict = {}
current = nested_dict
levels = ["level1", "level2", "level3", "level4"]
for i, key in enumerate(levels):
    if i == (len(levels)-1):
        #last level gets value
        current[key] = "deep value"
    else:
        #create a new dictionary and go deeper
        current[key] ={}
        current = current[key]
print(nested_dict)
```

```python
# Question 43: Create a dictionary with nested lists: {'matrix': [[1, 2], [3, 4]], 'vector': [5, 6, 7]}
data = {'matrix': [[1, 2], [3, 4]], 'vector': [5, 6, 7]}

# Question 44: Find sum of each nested list
result ={}
for key, value in data.items():
    if isinstance(value[0], list):
        result[key] = [sum(num) for num in value]
    else:
        result[key] = sum(value)
print(result)
```

```python
# Question 49: Create a dictionary with nested sets: {'groups': {{1, 2, 3}, {4, 5, 6}}, 'categories': {{'a', 'b'}, {'c', 'd'}}}

data = {'groups': [{1, 2, 3}, {4, 5, 6}],'categories': [{'a', 'b'}, {'c', 'd'}] }
# Question 50: Find union of all nested sets
result = {}
for key, sets in data.items():
    union_set = set()
    for s in sets:
        union_set |= s
        result[key] =union_set
print(result)



```
