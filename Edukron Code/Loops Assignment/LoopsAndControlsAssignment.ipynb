Rules:

A year is a leap year if it is divisible by 4
However, if it is divisible by 100, it is NOT a leap year
Exception: If it is also divisible by 400, then it IS a leap year
Examples:

2000: Leap year (divisible by 400)
1900: Not a leap year (divisible by 100 but not 400)
2024: Leap year (divisible by 4 but not 100)
Hint: Check conditions in order: first divisible by 400, then by 100, then by 4. Use modulo operator (%) to check divisibility.

```python
def is_leap_year(y) :
    if y % 400 == 0:
        return True
    elif y % 100 == 0 :
        return False
    elif y % 4 == 0 :
        return True
    else :
        return False

year = int(input("Enter Any Year"))
if is_leap_year(year):
    print(f"Entered Year {year}: is a Leap Year")
else :
    print(f"Entered Year {year}: is Not a Leap Year")
```

An Armstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits.
Examples:
153 = 1³ + 5³ + 3³ = 1 + 125 + 27 = 153 (Armstrong number)
9474 = 9⁴ + 4⁴ + 7⁴ + 4⁴ = 6561 + 256 + 2401 + 256 = 9474 (Armstrong number)

The Collatz conjecture is one of the most famous unsolved problems in mathematics. The rules are:

Start with any positive integer n
If n is even, divide it by 2
If n is odd, multiply it by 3 and add 1
Repeat until you reach 1
Write a program that:

Takes a number as input
Shows the complete sequence until it reaches 1
Counts how many steps it took
Example: Starting with 6: 6 → 3 → 10 → 5 → 16 → 8 → 4 → 2 → 1 (8 steps)

Hint: While n != 1: if n%2==0, n=n//2, else n=3*n+1. Count steps and print each value.

```python

# Collatz Conjecture Program

n = int(input("Enter a positive integer: "))
steps = 0

print("Collatz sequence:")

while n != 1:
    print(n, end=" → ")
    if n % 2 == 0:   # even
        n = n // 2
    else:            # odd
        n = 3 * n + 1
    steps += 1

print(1)  # final value
print("Total steps:", steps)
```

Write a program that implements a Caesar cipher to encrypt and decrypt messages.
The Caesar cipher shifts each letter by a fixed number of positions in the alphabet.
Example: With shift=3
'A' becomes 'D'
'Hello' becomes 'Khoor'
Your program should:

Ask for a message and shift value
Provide options to encrypt or decrypt
Handle both uppercase and lowercase letters
Keep non-alphabetic characters unchanged
Hint: Loop through each character. Use ord() to get ASCII value, shift it, use chr() to convert back. Use modulo 26 for wrapping. Check if char.isalpha().

```python
def ceasar_ciphar(text, shift, mode):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord("A") if char.isupper() else ord("a")# choose base depending on case
            if mode == "encrypt":
                result += chr((ord(char) - base + shift) % 26 +base)
            elif mode == "decrypt":
                result += chr((ord(char) - base - shift) % 26 + base)
            else:
                result += char
    return result

while True:
    print("\n~~~~~~~~Ceasar Ciphar~~~~~~")
    print("1---> Encrypt")
    print("2---> Decrypt")
    print("3---> Exit")

    choice = int(input("Enter Your Choice"))
    if choice == 3:
        print("Exiting program...gooodbye")
        break
    msg = input("Enter your messages")
    shift = int(input("Enter Shift Value"))
    if choice == 1:
        encrypted = ceasar_ciphar(msg, shift, "encrypt")
        print("Encrypted Messase", encrypted)

    elif choice == 2:
        decrypted = ceasar_ciphar(msg, shift, "decrypt")
        print("Decrypted Messase", decrypted)

    else :
        print("Give Correct choice(1, 2 or 3)")
```

```python
balance=10000
pin=1234
trans_history=[]
#verify pin
attempts=0
while attempts < 3:
    entered_pin= int(input("Enter your PIN number"))
    if entered_pin == pin:
        print("Pin accepted    Welcome")
        break
    else:
        attempts += 1
        print("Incorrect PIN   attempts left: " ,3 - attempts)
if attempts == 3:
    print("You have reached to maximum attemts Try after 24 hours!")
    exit()

while True:
    print("\n--- ATM Menu ---")
    print("1. Check Balance")
    print("2. Withdraw Money")
    print("3. Deposit Money")
    print("4. Change PIN")
    print("5. View Last 5 Transactions")
    print("6. Exit")

    choice = int(input("Enter your choice  (1-6)"))
    if choice == 1:
        print("Your Current Balance is: ", balance)
        trans_history.append("Balance Checked")
    elif choice == 2:
        try:
            withdraw_amt= float(input("Enter Withdraw Amount"))
            if withdraw_amt > balance:
                print("Insufficient Balance")
            elif withdraw_amt <=0:
                print("Invalid Amount")
            else:
                balance -= withdraw_amt
                print("Withdraw Successful.. New Balace is: ",balance)
                trans_history.append(f"Withdrew {withdraw_amt}")
        except ValueError:
            print("Invalid input Enter correct amount")
    elif choice == 3:
        try:
            deposit = float(input("Enter Deposit Amount"))
            if deposit <=0:
                print("Enter Valid Amount")
            else:
                balance += deposit
                print("Deposit Successful.. New Balance is:", balance)
                trans_history.append(f"Deposited {deposit}")
        except ValueError:
            print("Invalid input Enter correct deposit amount")
    elif choice == 4:
        new_pin = input("Emter New Pin:")
        if len(new_pin) == 4 and new_pin.isdigit():
            pin = new_pin
            print("PIN changed successful..")
            trans_history.append("Changed PIN")
        else:
            print("Enter Valid input formate")
    elif choice == 5:
        print("\n------ Last 5 Transactions--------")
        for t in trans_history[-5:]:
            print(t)
    elif choice == 6:
        print("Thank You for using this ATM...  Visit Again")
        break
    else:
        print("Invalid Choice.. Enter Valid Choice (1-6)")
```

```python
n = int(input("Enter a decimal number: "))

binary_digits = []   # list to store remainders (bits)

# Step 1: extract bits using modulo and division
while n > 0:
    remainder = n % 2
    binary_digits.append(str(remainder))  # store bit
    n //= 2

# Step 2: reverse the collected bits to form correct binary
binary_str = ""
for i in range(len(binary_digits) - 1, -1, -1):
    binary_str += binary_digits[i]

print("Binary representation =", binary_str)
```

Calculate compound interest for n years: A = P(1 + r/100)^t

Hint: Use for loop to multiply principal by (1 + r/100) t times. Or use: A = P * (1 + r/100)**t.

```python
P = float(input("Enter principal amount (P): "))
r = float(input("Enter annual interest rate (%): "))
t = int(input("Enter number of years (t): "))
A = P   # start with principal
# Step 1: loop t times, compounding each year
for year in range(t):
    A *= (1 + r / 100)

print(f"Amount after {t} years = {A:.2f}")
```

Problem 9.29: Perfect Numbers in Range (Hard)
Print all perfect numbers between 1 and n.
Hint: Outer loop for each number. For each, use inner loop to find sum of divisors (Problem 6.3). If sum equals number, print it.

```python
n = int(input("Enter a number: "))
print(f"Perfect numbers between 1 and {n} are:")
for num in range(2, n + 1):   # outer loop: check each number
    divisor_sum = 0
    for i in range(1, num):   # inner loop: sum proper divisors
        if num % i == 0:
            divisor_sum += i
    if divisor_sum == num:    # check if sum equals the number
        print(num)
```

Problem 9.30: Nested Factorial Sum (Hard)
Calculate: 1! + 2! + 3! + ... + n!
Hint: Outer loop from 1 to n+1. For each i, calculate factorial using inner loop (Problem 2.2), add to sum.

```python
n = int(input("Enter a number: "))
total_sum = 0
# Outer loop: go through each number from 1 to n
for num in range(1, n + 1):
    factorial = 1
    # Inner loop: calculate factorial of num
    for i in range(1, num + 1):
        factorial *= i
    # Add factorial to total sum
    total_sum += factorial
print(f"Sum of factorials up to {n}! = {total_sum}")
```

Print each digit of a number on a separate line.

```python
n = int(input("Enter a number: "))
digits = []   
while n > 0:
    digits.append(n % 10)  
    n //= 10                
for i in range(len(digits) - 1, -1, -1):
    print(digits[i])
```

Problem 10.15: Remove Specific Digit (Hard)
Remove all occurrences of a specific digit from a number.
Hint: Extract each digit. If digit != target, build new number: result = result * 10 + digit.

```python
n = int(input("Enter a number: "))
target = int(input("Enter digit to remove: "))

result = 0
multiplier = 1

while n > 0:
    digit = n % 10        # extract last digit
    if digit != target:   # keep only non-target digits
        result = digit * multiplier + result
        multiplier *= 10
    n //= 10              # remove last digit

print(f"Number after removing {target} = {result}")
```

Convert decimal to binary using while loop.

Hint: Same as Problem 9.24. While n > 0: binary = str(n%2) + binary, n = n//2.

```python
n = int(input("Enter a decimal number: "))

binary = ""   # start with empty string
while n > 0:
    binary = str(n % 2) + binary   # prepend remainder (bit)
    n //= 2                        # reduce number by integer division
print(f"Binary representation = {binary}")
```

Problem 10.19: Hexadecimal Conversion (Hard)
Convert decimal to hexadecimal using while loop.

Hint: hex_digits = '0123456789ABCDEF'. While n > 0: hex_str = hex_digits[n%16] + hex_str, n = n//16.

```python
n = int(input("Enter a decimal number: "))

hex_digits = "0123456789ABCDEF"
hex_str = ""   # start with empty string
while n > 0:
    remainder = n % 16
    hex_str = hex_digits[remainder] + hex_str   # prepend hex digit
    n //= 16                                    # reduce number
print(f"Hexadecimal representation = {hex_str}")
```
