```python
# Question 4: Sort the list [64, 34, 25, 12, 22, 11, 90] in descending order
numbers = [64, 34, 25, 12, 22, 11, 90]
print(numbers)
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] < numbers[j]:
            numbers[i],numbers[j] = numbers[j],numbers[i]
print(numbers)
```

```python
# Question 10: Create a list of prime numbers between 1 and 50
prime_number = []
for num in range(2,51):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_number.append(num)
print(prime_number)            
```

```python
# Question 31: Create a list of complex numbers: [1+2j, 3+4j, 5+6j]

complex_list =[]
real_part = 1
imag_part = 2
count = 3
i= 0
while i < count:
    z = complex(real_part, imag_part)
    complex_list.append(z)
    real_part += 2
    imag_part +=2
    i += 1
print(complex_list)
```

```python
# Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]
#def depth(lst):
data = [1, [2, 3], [4, [5, 6],[5,7]], 7] 

max_depth = 1
stack = [(data, 1)]  # (current_list, current_depth)

while stack:
    current_list, current_depth = stack.pop()
    max_depth = max(max_depth, current_depth)

    for item in current_list:
        if isinstance(item, list):
            stack.append((item, current_depth + 1))

print(max_depth)
```
